```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating COunt'})
```

```python
ratings
```

## Observation
1. When Rating is betwwen 4.5 to 4.9------> Excellent
2. When Ratingt are between 4.0 to 3.4------> Very good
3. When  Rating is between 3.5 to 3.9------> Good
4. When Rating is betwwen 3.0 to 3.4------> Average
5. When Rating is between 2.5 to 2.9------> Average
6. When rating is betwwen 2.0 to 2.4-----> Poor
